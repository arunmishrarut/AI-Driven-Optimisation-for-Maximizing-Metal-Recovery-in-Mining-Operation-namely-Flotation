```python
import pandas as pd
import numpy as np

# Assume data is already loaded and cleaned as before

# Convert 'date' to datetime and set as index
data['date'] = pd.to_datetime(data['date'])
data = data.set_index('date')

# Resample to hourly intervals and average all columns
data_hourly = data.resample('h').mean()

# Calculate change from previous row (temporary variables, not added as columns)
diff = data_hourly.diff()
starch_flow_change = diff['Starch Flow'] / data_hourly['Starch Flow'].shift(1)
amina_flow_change = diff['Amina Flow'] / data_hourly['Amina Flow'].shift(1)
ph_change = diff['Ore Pulp pH'] / data_hourly['Ore Pulp pH'].shift(1)

# Set your constraints using these temporary variables
constraints = [
    (
        data_hourly[
            [
                'Flotation Column 01 Air Flow',
                'Flotation Column 02 Air Flow',
                'Flotation Column 03 Air Flow',
                'Flotation Column 04 Air Flow',
                'Flotation Column 05 Air Flow',
                'Flotation Column 06 Air Flow',
                'Flotation Column 07 Air Flow'
            ]
        ].sum(axis=1) >= 1490,
        'Minimum sum of airflow for all columns'
    ),
    (
        (data_hourly['Starch Flow'] + data_hourly['Amina Flow']) >= 1000,
        'Min starch and Amina flow combined'
    ),
    (
        starch_flow_change.between(-0.05, 0.05),
        '%age starch Increase in 1 hour'
    ),
    (
        amina_flow_change.between(-0.05, 0.05),
        '%age Amina increase in 1 hour'
    ),
    (
        ph_change.between(-0.05, 0.05),
        '%age pH change'
    ),
    (
        data_hourly[
            [
                'Flotation Column 01 Air Flow',
                'Flotation Column 02 Air Flow',
                'Flotation Column 03 Air Flow',
                'Flotation Column 04 Air Flow',
                'Flotation Column 05 Air Flow',
                'Flotation Column 06 Air Flow',
                'Flotation Column 07 Air Flow'
            ]
        ].apply(lambda x: x.between(220, 350)).all(axis=1),
        'Airflow bound'
    ),
    (
        data_hourly[
            [
                'Flotation Column 01 Level',
                'Flotation Column 02 Level',
                'Flotation Column 03 Level',
                'Flotation Column 04 Level',
                'Flotation Column 05 Level',
                'Flotation Column 06 Level',
                'Flotation Column 07 Level'
            ]
        ].apply(lambda x: x.between(150, 750)).all(axis=1),
        'Level Bound'
    ),
    (
        data_hourly['Amina Flow'].between(250, 700),
        'Amina Bound'
    ),
    (
        data_hourly['Ore Pulp Flow'].between(370, 415),
        'Pulp Flow bound'
    ),
    (
        data_hourly['Ore Pulp pH'].between(8.75, 10.8),
        'pH bound'
    ),
    (
        data_hourly['Ore Pulp Density'].between(1.5, 1.85),
        'Density Bound'
    )
]

mask = np.ones(len(data_hourly), dtype=bool)
for condition, description in constraints:
    mask &= condition

# Only keep original columns
filtered_data = data_hourly.loc[mask, data.columns].copy()
```
